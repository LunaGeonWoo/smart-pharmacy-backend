```python
import pandas as pd

df = pd.read_excel("e약은요정보검색.xlsx")
df
```

```python
df = df.dropna(subset=['이 약의 효능은 무엇입니까?'])
df
```

```python
prompt = df.iloc[:,[4]]
prompt.columns = ["prompt"]
prompt = prompt.reset_index(drop=True)
prompt
```

```python
completion = pd.DataFrame(
    {
        "completion": [
            f"다음과 같은 성분을 추천합니다.\n주성분 : {name}\n"
            for serial, name in zip(df["품목일련번호"], df["제품명"])
        ]
    }
)
completion
```

```python
prompt = prompt.join(completion)
prompt
```

```python
prompt.to_json("prompt.jsonl", orient="records", lines=True, force_ascii=False)
```
